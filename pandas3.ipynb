```python
import pandas as pd

df = pd.read_csv("datasets/Minimum Wage Data.csv", encoding="latin")
# unicode decode error - pandas wants to use UTF8 encoding, but we have an encoding issue, therefore we use latin encoding instead
df.to_csv("datasets/minwage.csv", encoding="utf-8") 
# encode back to utf8 

df = pd.read_csv("datasets/minwage.csv") 
# read new csv file
df.head()
```

```python
gb = df.groupby("State")
# creates groupby object
gb.get_group("Alabama").set_index("Year").head()
# get the Alabama group
```

```python
act_min_wage = pd.DataFrame()

for name, group in df.groupby("State"): 
    if act_min_wage.empty: 
        act_min_wage = group.set_index("Year")[["Department.Of.Labor.Cleaned.Low.Value.2020.Dollars"]].rename(columns={"Department.Of.Labor.Cleaned.Low.Value.2020.Dollars": name})
        # iterating over all groups, renaming them - this gives us DoL min wage in a column sorted by year, with States as headings
    else: 
        act_min_wage = act_min_wage.join(group.set_index("Year")[["Department.Of.Labor.Cleaned.Low.Value.2020.Dollars"]].rename(columns={"Department.Of.Labor.Cleaned.Low.Value.2020.Dollars": name}))
        
act_min_wage.head()

# this is very common, so it has a build in pandas function
        
    
```

```python
issue_df = df[df["Department.Of.Labor.Cleaned.Low.Value.2020.Dollars"]==0]
issue_df.head()
# identifying problems in dataset
```

```python
issue_df["State"].unique()
# shows which states are the problems, useful if we have a lot of missing data. No reason for us to continue with data from these states
```

```python
import numpy as np 

act_min_wage.replace(0, np.NaN).dropna(axis=1).corr().head()
# replace zeroes with NaN, axis =1 will get rid of a column. axis =0 is rows
# we now have minimum wage correlation data
```

```python
min_wage_corr = act_min_wage.replace(0, np.NaN).dropna(axis=1).corr().head()

# replaces all missing zero data with NaN, and removes the collumns

for problem in issue_df['State'].unique(): 
    if problem in min_wage_corr.columns:
        print("we're missing something here...")
        # shows if there is still a problem
    
```

```python
# what if a state didn't have a min wage and got one later? 
# currently we are assuming all zeroes are mistakes, but we need to explore more. 
grouped_issues = issue_df.groupby("State")

grouped_issues.get_group("Alabama")["Department.Of.Labor.Cleaned.Low.Value.2020.Dollars"].sum()
# if zero returns this means we never get minimum wage data for the state. 
```

```python
# we can iterate over all states
for state, data in grouped_issues: 
    if data["Department.Of.Labor.Cleaned.Low.Value.2020.Dollars"].sum() != 0.0: 
        print("We missed something!")
# if nothing prints, we never get any data for any of the states. 
```
